### What Is a Transformer?

A Transformer is a neural network architecture that processes sequences —
text, code, audio frames, pixel patches — by learning which parts of the
input to pay attention to when producing each output. It was introduced in
the 2017 paper *"Attention Is All You Need"* by Vaswani et al. and has since
become the foundation of nearly every large language model in existence.

The key insight is a mechanism called **self-attention**: instead of reading
a sequence token by token (like an RNN), the Transformer looks at *all tokens
simultaneously* and learns a weighted relationship between every pair. This
makes it highly parallelisable and exceptionally good at capturing long-range
dependencies in text.

### Why Build It From Scratch?

Using a library like HuggingFace hides every interesting decision behind an
API call. Building from scratch means you will see exactly:

- Why Q, K, V exist and where they come from
- What the square-root scaling in attention actually prevents
- How layer normalisation stabilises training
- Why residual connections are not optional
- How a language model head turns vectors into probabilities over words
